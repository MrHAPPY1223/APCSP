```python
#Create a dataframe object for your data here

data = pd.read_csv("https://data.cdc.gov/api/views/9j2v-jamp/rows.csv?accessType=DOWNLOAD")

green_diamond = dict(markerfacecolor='g', marker='D')
fig2, ax2 = plt.subplots()
ax2.set_title('Changed Outlier Symbols')
ax2.boxplot(data.YEAR, flierprops=green_diamond)
```

```python
plt.hist(data.ESTIMATE, bins=50)
plt.ylabel('Amount of Deaths per Year!')
plt.xlabel('Estimated Deaths Per Year!')
plt.title('Estimating Death!!')
plt.show()
```
